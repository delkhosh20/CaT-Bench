```python
import json
import os
```

```python
folder_path = 'gemini-1.5-flash-8b'

total_questions = 0
total_correct = 0
```

```python
for filename in os.listdir(folder_path):
    #print(filename)
    file_path = os.path.join(folder_path, filename)
    with open(file_path, 'r') as f:
        data = json.load(f)
    #print(data)
    
    file_total = len(data)
    file_correct = 0
    incorrect_entries = []

    for entry in data:
        answer = entry['answer'].strip().lower()
        if 'no' in answer:
            file_correct += 1
        else:
            incorrect_entries.append({
                'goal': entry['goal'],
                'question': entry['question'],
                'answer': entry['answer']
            })

    file_accuracy = (file_correct / file_total) * 100 if file_total > 0 else 0
    print(f"\nFile: {filename}")
    print(f"  Total questions: {file_total}")
    print(f"  Correct (NO) answers: {file_correct}")
    print(f"  Accuracy: {file_accuracy:.2f}%")

    if incorrect_entries:
        print(f"  Incorrectly answered questions:")
        for i, entry in enumerate(incorrect_entries, 1):
            print(f"    {i}. Goal: {entry['goal']}")
            print(f"       Question: {entry['question']}")
            print(f"       Answer: {entry['answer']}")

    total_questions += file_total
    total_correct += file_correct
```

```python
overall_accuracy = (total_correct / total_questions) * 100 if total_questions > 0 else 0

print(f"\n=== Overall Summary ===")
print(f"Total files processed: {len([f for f in os.listdir(folder_path) if f.endswith('.json')])}")
print(f"Total questions: {total_questions}")
print(f"Total correct (NO) answers: {total_correct}")
print(f"Overall accuracy: {overall_accuracy:.2f}%")
```
